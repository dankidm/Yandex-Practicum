# Телеком описание задачи

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

> Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

# Обучение моделей и подбор параметров

> Деление выборки

```python
X = data.drop(['Leave', 'customerID', 'BeginDate', 'EndDate'], axis=1)
y = data['Leave']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=RANDOM_STATE)
```

> Создание списков названий категориальных и количественных признаков 

```python
X['SeniorCitizen'] = X['SeniorCitizen'].astype('object')

num_cols = X.select_dtypes(['int', 'float']).columns
cat_cols = X.select_dtypes('object').columns
```

```python
cat_cols
```

> Создание обьекта ColumnTransformer для предобработки данных в Pipeline

```python
numeric = Pipeline([
    ('scaler', StandardScaler())
])

categoric = Pipeline([
    ('encoder', OneHotEncoder(drop='first'))
])

preprocessor = ColumnTransformer([
    ('num', numeric, num_cols),
    ('cat', categoric, cat_cols)
])
```

## LinearRegression

```python
lin_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

lin_pipe.fit(X_train, y_train)
lin_pred = lin_pipe.predict(X_train)
             
print('Значение метрики roc-auc на обучающей выборке модели LinearRegression', roc_auc_score(y_train, lin_pred))
```

## CatBoostClassifier

> Так как подбор параметров производится долго, я закомментировал код с перебором параметров и оставил в модели лучшие параметры

```python
cat_features=['Type', 'PaperlessBilling', 'PaymentMethod', 'gender', 'Partner', 
              'Dependents', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
              'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'MultipleLines']

cat_model = CatBoostClassifier(verbose=False, eval_metric='AUC', random_state=RANDOM_STATE)

cat_param = {
    'max_depth': [5],
    'iterations': [900],
    'learning_rate': [0.2]
}

cat_grid = GridSearchCV(cat_model, param_grid=cat_param, cv=4, scoring='roc_auc')

cat_grid.fit(X_train, y_train, cat_features=cat_features)

print('Значение метрики roc-auc на обучающей выборке модели CatBoostClassifier', cat_grid.best_score_)
```

## LGBMClassifier

```python
lgbm_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LGBMClassifier(random_state=RANDOM_STATE, learning_rate=0.1, max_depth=5, n_estimators=250))
])

lgbm_param = {
    'model__max_depth': [5],
    'model__learning_rate': [0.1],
    'model__n_estimators': [250]
}

lgbm_grid = GridSearchCV(lgbm_pipe, param_grid=lgbm_param, cv=4, scoring='roc_auc')

lgbm_grid.fit(X_train, y_train)

lgbm_pred = lgbm_grid.predict_proba(X_train)[:, 1]
             
print('Значение метрики roc-auc на обучающей выборке модели LGBMClassifier', lgbm_grid.best_score_)
```

# Валидация модели

```python
test_pred = cat_grid.predict_proba(X_test)[:, 1]
print('Значение метрики roc-auc на тестовой выборке лучшей модели (CatBoostClassifier)', roc_auc_score(y_test, test_pred))
```

# Анализ важности признаков

```python
importance = cat_grid.best_estimator_.get_feature_importance()

sorted_idx = importance.argsort()

plt.figure(figsize=(15, 12))
plt.barh(range(X.shape[1]), importance[sorted_idx])
plt.yticks(range(X.shape[1]), X.columns[sorted_idx])
plt.title("Feature Importance")
plt.show()
```

> Самым важным параметром для лучшей модели оказалось количество дней которое абонент пользовался услугами компании

# Создание графика ROC кривой

```python
sns.set(font_scale=1.5)
sns.set_color_codes("muted")

fpr, tpr, threshold = roc_curve(y_test, test_pred)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(15, 12))
plt.title('ROC curve')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# Визуализация матрицы ошибок

```python
matrix = confusion_matrix(y_test, cat_grid.predict(X_test))

ax = sns.heatmap(matrix/np.sum(matrix), annot=True, fmt='.2%', cmap='Blues')

ax.set_title('Confusion Matrix\n\n');
ax.set_xlabel('\nПредсказанное значение')
ax.set_ylabel('Реальное значение ');

ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

plt.show()
```
